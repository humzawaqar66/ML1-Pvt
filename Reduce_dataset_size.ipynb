```python
import keras
import random
import collections
import numpy as np
import matplotlib.pyplot as plt

import tensorflow
def display_random_images_with_labels(d, n=10):
    """
        accept dataset in tuple form, first index
        is np arrays (images) and second is labels
        and display n random images with labels
        from it
        Inputs:
            d (tuple): dataset images, dataset labels on same indexs
            n (int): number of samples to display (default: 10)
        Output:
            None
    """
    choices = list(range(len(d[0])))
    for i in range(n):
        index = random.choice(choices)
        choices.remove(index)
        print("index:",index)
        print("Lable:",d[1][index])
        plt.imshow(d[0][index], cmap='gray')
        plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
def reduce_dataset_size(images, labels, target_size):
    # Split the dataset into training and testing sets
    _, images_temp, _, labels_temp = train_test_split(images, labels, test_size=target_size, stratify=labels, random_state=42)
    return images_temp, labels_temp
# Load the MNIST dataset
dataset = tensorflow.keras.datasets.mnist.load_data()
# Extract the training dataset
train_images, train_labels = dataset[0]
# Reduce the size of the training dataset to 6,000 samples
target_size = 6000
reduced_train_images, reduced_train_labels = reduce_dataset_size(train_images, train_labels, target_size)
# Display the information about the reduced dataset
print("Reduced dataset size:", len(reduced_train_images))
print("Number of unique classes in the reduced dataset:", len(set(reduced_train_labels)))
```
