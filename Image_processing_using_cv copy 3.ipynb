```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path = "IMG_01.jpeg"
```

```python
img=cv2.imread(path)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('IMG_01')

```

```python
import numpy as np
directory_path = r'/Users/hwaqar/Desktop/images'
if img is None:
    print("Error: Unable to load the image.")
else:
    height, width, _ = img.shape
    # Create a window to display the original image
    # Flip the image in every angle (360 degrees)
    for angle in range(0, 360, 1):
        # Rotate the image by the current angle
        rotated_image = cv2.warpAffine(img, cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1), (width, height))
        # Display the rotated image
        save_path = f'{directory_path}\\image_{angle}.jpeg'
        cv2.imwrite(save_path, rotated_image)
    cv2.destroyAllWindows()

```

```python
img_1=cv2.rectangle(img,(110,30),(500,220),(0,0,255),2)
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB))
plt.title('Rectangle')
```

```python
blur = cv2.GaussianBlur(img, (35,35),0)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title('blur')
```

```python
img_1=img+np.array([100])
brightened_image = np.clip(img_1, 0, 255).astype(np.uint8)
plt.imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
plt.title('Brightened Image')
```
