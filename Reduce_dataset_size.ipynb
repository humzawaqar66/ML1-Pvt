```python
## runnning code
def display_random_images_with_labels(d, n=10):
    """
    Display n random images with labels from the dataset.
    Inputs:
        d (tuple): dataset images, dataset labels on the same indices
        n (int): number of samples to display (default: 10)
    Output:
        None
    """
    choices = list(range(len(d[0])))
    for i in range(n):
        index = random.choice(choices)
        choices.remove(index)
        label = d[1][index]
        print("Index:", index)
        print("Label:", label)
        plt.imshow(d[0][index], cmap='gray')
        plt.title(f"Label: {label}")
        plt.show()

# Loading dataset
full_dataset = tf.keras.datasets.mnist.load_data()

# Number of samples to compress
compressed_train_size = 6000
compressed_test_size = 1000

# Randomly shuffle the indices for each label
indices_per_label = collections.defaultdict(list)
for i, label in enumerate(full_dataset[0][1]):
    indices_per_label[label].append(i)

# Select a subset of samples for each label for the training dataset
compressed_train_indices = []
for label_indices in indices_per_label.values():
    compressed_train_indices.extend(random.sample(label_indices, min(compressed_train_size // len(indices_per_label), len(label_indices))))

# Shuffle the selected indices to ensure randomness
random.shuffle(compressed_train_indices)

# Create the compressed training dataset
compressed_train_dataset = (
    full_dataset[0][0][compressed_train_indices],
    full_dataset[0][1][compressed_train_indices]
)
# Select a subset of samples for each label for the test dataset
compressed_test_indices = []
for label_indices in indices_per_label.values():
    compressed_test_indices.extend(random.sample(label_indices, min(compressed_test_size // len(indices_per_label), len(label_indices))))

# Shuffle the selected indices to ensure randomness
random.shuffle(compressed_test_indices)

# Create the compressed testing dataset
compressed_test_dataset = (
    full_dataset[0][0][compressed_test_indices],
    full_dataset[0][1][compressed_test_indices]
)

# Description of compressed datasets
print("There are", len(compressed_train_dataset[0]), "images in compressed training dataset")
print("There are", len(compressed_test_dataset[0]), "images in compressed testing dataset")
print("___________________\n\n")

# Randomly printing 10 images with labels from compressed training dataset
print("Randomly printing 10 images with labels from compressed training dataset")
display_random_images_with_labels(compressed_train_dataset)
print("___________________\n\n")

# Randomly printing 10 images with labels from compressed testing dataset
print("Randomly printing 10 images with labels from compressed testing dataset")
display_random_images_with_labels(compressed_test_dataset)
print("___________________\n\n")

# Number of labels in the compressed training and testing datasets
print("There are", len(set(compressed_train_dataset[1])), "unique classes in compressed training dataset")
print("Breakdown of each label is below (format: dict key is label, dict value is occurrence of that label/number of images for that label)")
display(collections.Counter(compressed_train_dataset[1]))

print("There are", len(set(compressed_test_dataset[1])), "unique classes in compressed testing dataset")
print("Breakdown of each label is below (format: dict key is label, dict value is occurrence of that label/number of images for that label)")
display(collections.Counter(compressed_test_dataset[1]))
print("___________________\n\n")
```
